# 🛒E-Commerce Profit Analysis

Exploratory analysis of e-commerce sales, profit, and margin patterns.

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML

plt.style.use('ggplot')
```

## 2. Load Dataset

```python
df = pd.read_csv('data/ecommerce_sales_data.csv')
df.head()
```

## 3. Basic Inspection

```python
display(df.tail())
print('Shape:', df.shape)
print(df.columns.tolist())
print(df.dtypes)
print('Missing values:\n', df.isna().sum())
print('Duplicate rows:', df.duplicated().sum())
display(df.describe(include='all').T)
```

## 4. Cleaning

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
for col in ['Sales', 'Profit']:
    df[col] = pd.to_numeric(df[col].astype(str).str.replace(r'[\$,]', '', regex=True).str.strip(), errors='coerce')
df['Profit Margin'] = df['Profit'] / df['Sales']
df['Year'] = df['Order Date'].dt.year

clean_df = df.drop_duplicates().copy()
```

## 5. Summary Statistics

```python
clean_df[['Sales','Profit','Quantity','Profit Margin']].describe().T
```

## 6. Correlation Heatmap

```python
corr = clean_df[['Sales','Profit','Quantity','Profit Margin']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

## 7. Yearly Sales and Profit Trend

```python
yearly = clean_df.groupby('Year')[['Sales','Profit']].sum()
plt.figure(figsize=(8,5))
plt.plot(yearly.index.astype(str), yearly['Sales'], marker='o')
plt.plot(yearly.index.astype(str), yearly['Profit'], marker='o')
plt.title('Yearly Sales and Profit Trend')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend(['Sales','Profit'])
plt.show()
```

## 8. Profit Margin by Category

```python
category = clean_df.groupby('Category')[['Sales','Profit']].sum()
category['Profit Margin'] = category['Profit'] / category['Sales']
category = category.sort_values('Profit Margin')
plt.figure(figsize=(7,5))
plt.bar(category.index, category['Profit Margin'])
plt.title('Profit Margin by Category')
plt.xlabel('Category')
plt.ylabel('Profit Margin')
plt.show()
```

## 9. Profit Margin by Region

```python
region = clean_df.groupby('Region')[['Sales','Profit']].sum()
region['Profit Margin'] = region['Profit'] / region['Sales']
region = region.sort_values('Profit Margin')
plt.figure(figsize=(7,5))
plt.bar(region.index, region['Profit Margin'])
plt.title('Profit Margin by Region')
plt.xlabel('Region')
plt.ylabel('Profit Margin')
plt.show()
```

## 10. Profit Margin by Product

```python
product = clean_df.groupby('Product Name')[['Sales','Profit']].sum()
product['Profit Margin'] = product['Profit'] / product['Sales']
product = product.sort_values('Profit Margin')
plt.figure(figsize=(10,6))
plt.bar(product.index, product['Profit Margin'])
plt.title('Profit Margin by Product')
plt.xlabel('Product')
plt.ylabel('Profit Margin')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## 11. Problem Statement

Strong sales do not always translate into equally strong profit. The analysis focuses on where profitability is weaker and what business actions can improve margin performance.

## 12. Recommendations

- Review pricing and cost structure for low-margin products.
- Prioritize high-margin products in promotions.
- Track margin monthly by product, category, and region.
- Use profit margin as a decision metric, not sales alone.
